```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
points  =100 
phi = np.linspace(-np.pi,np.pi,points)
x0 = 3 * np.cos(phi)
y0 = 2 * np.sin(phi)

def rotate(arr,n=3):
    angles = np.array([np.pi*2 /n * i for i in range(1,n)])
    
    return [
        np.dot(np.array([[np.cos(a), np.sin(a)],[-np.sin(a),np.cos(a)]]),arr) for a in angles
    ]
shape = np.array([[x0],[y0]]).shape
others = rotate(np.array([[x0],[y0]]).reshape(shape[0],shape[2]),n=3)    

plt.plot(x0,y0)
for o in others:
    plt.plot(o[0],o[1])
plt.grid()
```
